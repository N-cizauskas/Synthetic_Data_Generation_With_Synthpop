```python
# packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dfs
between_sex = pd.read_csv('between_sex.csv')
between_age = pd.read_csv('between_age.csv')
melted_between_sex = pd.read_csv('melted_between_sex.csv')
melted_between_age = pd.read_csv('melted_between_age.csv')

# rename data column for graphing titles
melted_between_sex = melted_between_sex.rename(columns={'data': 'Data Source'})
melted_between_age = melted_between_age.rename(columns={'data': 'Data Source'})
```

```python
# graph a boxplot to show the comparison between data sources and types, with outliers and median
fig = sns.boxplot(
    data=melted_between_age,
    x="variable", y="value", hue="Data Source"
)
sns.despine(offset=10, trim=True)
fig.set(xlabel='Data Types', ylabel='Values')
```

```python



# Draw a nested violinplot
# This shows the distributions of each df
sns.set_theme(style="dark")

fig2 = sns.violinplot(
    data=melted_between_age,
    x="variable", y="value", hue="Data Source",
               split=True, fill=False,
               palette={"real": "g", "synthetic": ".35"})
fig2.set(xlabel='Data Types', ylabel='Values')
```
